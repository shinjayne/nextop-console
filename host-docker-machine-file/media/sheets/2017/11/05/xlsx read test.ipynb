```python
def fill_dict(df,dict):
    months = sorted(list(set(df["년월"])))
    for m in months:
        datestr = str(m)[:10]
        sum = df[df["년월"]==m]["투입수량"].sum()
        dict[datestr] = sum 
        
    return None
```

```python
dflist = [df2010, df2011,df2012, df2013, df2014, df2015, df2016 , df2017]

def make_company_list(dflist):
    result=set()
    for df in dflist:
        result = set(list(result) + list(set(df["업체명"])))
    
    return list(result)

def make_company_yearly_chart(dflist, companies):
    result = [["회사","2010","2011","2012","2013","2014","2015","2016","2017"]]
    for c in companies:
        row = [str(c)]
        for df in dflist: 
            sum = df[df["업체명"]==c]["투입수량"].sum()
            row.append(sum)
            
        result.append(row)
    return result 
```
